## Importing The Dataset

```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "creditcard_2023.csv"

# Load the latest version
data = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "nelgiriyewithana/credit-card-fraud-detection-dataset-2023",
  file_path,
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:", data.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Feature-by-Feature Analysis

```python
for col in data.columns:
    print(f"\n*** Feature: {col} ***")
    print(" - Type:", data[col].dtype)
    print(" - Unique values:", data[col].nunique())

    if np.issubdtype(data[col].dtype, np.number):
        print(" - Min, Max, Mean, Std:", data[col].min(), data[col].max(), data[col].mean(), data[col].std())
        sns.histplot(data[col], bins=50, kde=True)
        plt.title(f"Distribution of {col}")
        plt.show()
    else:
        print(" - Value counts:\n", data[col].value_counts())
```

## Class Imbalance Analysis

```python
fraud_counts = data['Class'].value_counts()
print(fraud_counts)

sns.countplot(x='Class', data=data)
plt.title('Class Distribution (0 = Non-Fraud, 1 = Fraud)')
plt.show()

fraud_ratio = fraud_counts[1] / fraud_counts.sum()
print(f"Fraudulent Transaction Ratio: {fraud_ratio:.4f}")
```

## Correlation Analysis

```python
corr = data.corr()

corr_with_class = corr['Class'].drop('Class').sort_values(ascending=False)
print("Top features correlated with Class:\n", corr_with_class.head(10))
print("Lowest correlated features with Class:\n", corr_with_class.tail(10))

plt.figure(figsize=(10, 8))
sns.heatmap(data[corr_with_class.head(10).index.tolist() + ['Class']].corr(),
            annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix: Top Features vs Class')
plt.show()
```

## Feature Distributions by Class

```python
for feature in ['Amount'] + corr_with_class.head(5).index.tolist():
    plt.figure(figsize=(10, 4))
    sns.kdeplot(data=data[data['Class']==0][feature].replace([np.inf, -np.inf], np.nan).dropna(),
                label='Non-Fraud', fill=True)
    sns.kdeplot(data=data[data['Class']==1][feature].replace([np.inf, -np.inf], np.nan).dropna(),
                label='Fraud', fill=True)
    plt.title(f"Density Plot of {feature} by Class")
    plt.legend()
    plt.show()
```

## Outlier Detection using Box Plot

```python
features = data.columns.difference(['Class'])

for feature in features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='Class', y=feature, data=data)
    plt.title(f'Boxplot of {feature} by Class')
    plt.yscale('log') if feature == 'Amount' else None
    plt.show()

```

## Applying PCA for Dimensionality Reduction

```python
from sklearn.decomposition import PCA
X = data.drop(columns=['Class' , 'Amount'])
y = data['Class']
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X)
pca_features = pd.DataFrame(
    X_pca,
    columns=[f'PC{i+1}' for i in range(5)]
)
X_final_features = pd.concat([pca_features, data['Amount']], axis=1)
print("Final features shape:", X_final_features.shape)
X_final_features.head()
```

## Splitting Data for Training and Testing and Scaling Features using Standard Scalar

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
scaler = StandardScaler()
X_train,X_test,y_train,y_test = train_test_split(X_final_features,y,test_size=0.25,random_state=42,stratify=y)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## XGBoost Model Training using hyperparameter tuning (RandomizedSearchCV) and evaluation with Confusion Matrix, Classification Report, Auccuracy & ROC curve:

```python
!pip install xgboost
```

```python
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
xgb_clf = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
param_dist = {
    'n_estimators': [100, 200, 300, 400],
    'max_depth': [3, 5, 7, 9, 11],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.7, 0.8, 0.9, 1.0],
    'gamma': [0, 1, 5]
}
xgb_random = RandomizedSearchCV(
    estimator=xgb_clf,
    param_distributions=param_dist,
    n_iter=20,
    scoring='accuracy',
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1
)
print("Training XGBoost with hyperparameter tuning...")
xgb_random.fit(X_train_scaled, y_train)
```

## Performance Metrics

```python
best_xgb = xgb_random.best_estimator_
print("Best Parameters:", xgb_random.best_params_)
y_pred = best_xgb.predict(X_test_scaled)
y_pred_proba = best_xgb.predict_proba(X_test_scaled)[:, 1]
print("\nAccuracy: \n", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print(f"\n {confusion_matrix(y_test,y_pred)} \n")
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("XGBoost Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

## ROC AUC Curve

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, label=f'XGBoost (AUC = {roc_auc:.2f})')
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("XGBoost ROC Curve")
plt.legend()
plt.show()
```

## Precision–Recall curve

```python
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.model_selection import learning_curve
import numpy as np

y_scores = best_xgb.predict_proba(X_test_scaled)[:, 1]
precision, recall, thresholds = precision_recall_curve(y_test, y_scores)
avg_precision = average_precision_score(y_test, y_scores)

plt.figure(figsize=(7,5))
plt.plot(recall, precision, label=f'XGBoost (AP = {avg_precision:.2f})')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()
```

## Learning Curve

```python
train_sizes, train_scores, test_scores = learning_curve(
    best_xgb, X_train_scaled, y_train, cv=5, scoring="accuracy", n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 10), random_state=42
)

train_scores_mean = np.mean(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)

plt.figure(figsize=(7,5))
plt.plot(train_sizes, train_scores_mean, 'o-', label="Training score")
plt.plot(train_sizes, test_scores_mean, 'o-', label="Cross-validation score")
plt.xlabel("Training examples")
plt.ylabel("Accuracy")
plt.title("Learning Curve for XGBoost")
plt.legend(loc="best")
plt.show()
```

## Quantum Machine Learing (Using Qiskit)

```python
from qiskit_aer import Aer
from qiskit.circuit.library import ZZFeatureMap
from qiskit_machine_learning.kernels import FidelityQuantumKernel
from qiskit_machine_learning.algorithms.classifiers import QSVC
from qiskit.primitives import Sampler, Estimator
import qiskit
```

```python
feature_map = ZZFeatureMap(feature_dimension=6, reps=2, entanglement="full")
```

```python
qkernel = FidelityQuantumKernel(feature_map=feature_map)
```

```python
qsvc = QSVC(quantum_kernel=qkernel)
```

```python
print("Training QSVC...")
qsvc.fit(X_train_scaled, y_train)
print("Done.")
```

```python
y_pred = qsvc.predict(X_test_scaled)
scores = qsvc.decision_function(X_test_scaled)
proba = (scores - scores.min()) / (scores.max() - scores.min() + 1e-12)
```

```python
from sklearn.metrics import roc_auc_score
acc = accuracy_score(y_test, y_pred)
auc = roc_auc_score(y_test, proba)
print(f"Accuracy: {acc:.4f}")
print(f"\nClassification Report:\n{classification_report(y_test, y_pred, digits=4)}")
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Pred 0","Pred 1"], yticklabels=["True 0","True 1"])
plt.title(f"Confusion Matrix (reps=2, ent='full')")
plt.ylabel("True")
plt.xlabel("Predicted")
plt.show()
# ROC Curve
fpr, tpr, _ = roc_curve(y_test, proba)
plt.figure(figsize=(5,4))
plt.plot(fpr, tpr, label=f"AUC={auc:.4f}")
plt.plot([0,1],[0,1],'--',color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"ROC Curve (reps=2, ent='full')")
plt.legend()
plt.show()
```
